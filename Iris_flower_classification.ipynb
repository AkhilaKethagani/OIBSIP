```python
features=df2[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
target=df2['Species']
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
ypred=model.predict(x_test)
ypred
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
scored=r2_score(y_test,ypred)
print(scored)
```

```python
plt.scatter(y_test,ypred,c='b')
plt.plot([min(y_test),max(y_test)],[min(y_test),max(y_test)],c='r')
plt.xlabel('actual_values')
plt.ylabel('predicted values')
plt.show()
```
