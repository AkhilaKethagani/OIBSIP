```python
df[['state','Estimated Unemployment Rate (%)']]
```

```python
df['month']=df['Date'].dt.month
df
```

```python
df1=df.groupby('region').get_group('South')
df1
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='state',y='Estimated Unemployment Rate (%)',data=df1,hue='month',width=1)
plt.xticks(rotation=90)
plt.title('unemployed rate during covid 19 in Southern states of india',size=20,color='g')
plt.show()
```

```python
df2=df.groupby('region').get_group('North')
plt.figure(figsize=(5,5))
sns.scatterplot(x='state',y='Estimated Unemployment Rate (%)',data=df2,hue='month')
plt.xticks(rotation=90)
plt.title('unemployed rate during covid 19 in Northern states of india',size=20,color='g')
plt.show()
```

```python
df3=df.groupby('region').get_group('Northeast')
plt.figure(figsize=(10,5))
sns.barplot(x='state',y='Estimated Unemployment Rate (%)',data=df3,hue='month')
plt.xticks(rotation=90)
plt.title('unemployed rate during covid 19 in NorthEastern states of india',size=20,color='g')
plt.show()
```

```python
df4=df.groupby('region').get_group('East')
plt.figure(figsize=(10,5))
sns.barplot(x='state',y='Estimated Unemployment Rate (%)',data=df4,hue='month')
plt.xticks(rotation=90)
plt.title('unemployed rate during covid 19 in Eastern states of india',size=20,color='g')
plt.show()
```

```python
df5=df.groupby('region').get_group('West')
plt.figure(figsize=(10,5))
sns.barplot(x='state',y='Estimated Unemployment Rate (%)',data=df5,hue='month')
plt.xticks(rotation=90)
plt.title('unemployed rate during covid 19 in wastern states of india',size=20,color='g')
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(x='state',y='Estimated Employed',data=df.sort_values('Estimated Employed',ascending=False),hue='region')
plt.xticks(rotation=90)
plt.title('employed status in india',size=20)
plt.show()
```

```python
plt.figure(figsize=(15,7))
sns.scatterplot(x='state',y='Estimated Labour Participation Rate (%)',data=df,hue='region')
plt.xticks(rotation=90)
plt.title('Estimated Labour participation rate in india',size=20)
plt.show()
```
